```python
import pandas as pd
from scipy import stats
```

```python
Education = pd.read_csv("C:/Users/jenni/Desktop/DataScience/Final Project/FinalDataset2.csv")
```

```python
Education.head()
```

```python
education_crosstab = pd.crosstab(Education['Region'], Education['DegreeGrantingYN'])
```

```python
education_crosstab
```

```python
stats.chi2_contingency(education_crosstab)
```

#Looks like since the p-value is so low/significant that degree granting status and region is related. The assumptions are met for this analysis because all the sets are over 5. Let's now look at Degree of Urbanization.

```python
education_crosstab1 = pd.crosstab(Education['DegreeOfUrban'], Education['DegreeGrantingYN'])
```

```python
education_crosstab1
```

```python
stats.chi2_contingency(education_crosstab1)
```

#Looks like since the p-value is so low/significant that degree granting status and degree of urbanization is related. The assumptions are not met for this analysis because only some of the sets are over 5.
